```python
# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load the data 
data = pd.read_csv("/home/herz/Desktop/wind/Turbine_Data_Penmanshiel_11_2021-01-01_-_2021-07-01_1051.csv", sep=",", skiprows=9)

# Analyzing Wind Speed - Calculate statistics and plot distribution
wind_speed = data['Wind speed (m/s)']
wind_speed_stats = {
    'mean': wind_speed.mean(),
    'min': wind_speed.min(), 
    'max': wind_speed.max()
}

print("Wind Speed Statistics:")
print(f"Mean: {wind_speed_stats['mean']:.2f} m/s")  
print(f"Minimum: {wind_speed_stats['min']:.2f} m/s")
print(f"Maximum: {wind_speed_stats['max']:.2f} m/s")

# Plot distribution of wind speeds
plt.figure(figsize=(10, 6))
sns.histplot(wind_speed, bins=30, kde=True)
plt.title('Distribution of Wind Speeds') 
plt.xlabel('Wind Speed (m/s)')
plt.ylabel('Frequency')
plt.show()

# Relationship between wind speed and energy production
energy_production = data['Energy Export (kWh)']
plt.figure(figsize=(10, 6))
plt.scatter(wind_speed, energy_production, alpha=0.5)
plt.title('Wind Speed vs Energy Production')
plt.xlabel('Wind Speed (m/s)')
plt.ylabel('Energy Production (kWh)')
plt.show()

# Wind direction analysis
wind_direction = data['Wind direction (°)']

# Convert to radians for polar plot
wind_direction_rad = np.radians(wind_direction)

# Plot wind rose
plt.figure(figsize=(8, 8))
ax = plt.subplot(polar=True)
bins = np.arange(0, 2*np.pi, np.pi/8) 
counts, _, _ = plt.hist(wind_direction_rad, bins=bins)
ax.set_theta_zero_location("N")
ax.set_theta_direction(-1)
plt.title('Wind Rose - Wind Direction Frequency')
plt.show()
```

1. Distribution of Wind Speeds:
   - The histogram shows that the most frequent wind speed range is around 12-13 m/s, as indicated by the highest peak in the distribution.
   - The distribution appears to be slightly right-skewed, with a longer tail towards higher wind speeds.
   - The wind speeds range from approximately 8 m/s to 15 m/s, with a few instances of higher speeds up to around 20 m/s.
   - The KDE line suggests a relatively smooth distribution of wind speeds, with a clear peak and gradual decline on both sides.

2. Wind Speed vs Energy Production:
   - The scatter plot indicates a positive correlation between wind speed and energy production.
   - As the wind speed increases, there is a general trend of increasing energy production.
   - The relationship appears to be non-linear, with energy production increasing more rapidly at higher wind speeds.
   - There is some variability in energy production at similar wind speeds, as shown by the vertical spread of points at each wind speed level.
   - The plot suggests that higher wind speeds contribute significantly to energy production, with a notable increase in production beyond wind speeds of around 12 m/s.

3. Wind Rose - Wind Direction Frequency:
   - The wind rose plot reveals the frequency distribution of wind directions at the turbine location.
   - The predominant wind direction appears to be from the southwest (SW), as indicated by the largest blue segment in that direction.
   - The second most frequent wind direction is from the northeast (NE), shown by the second-largest segment.
   - Other notable wind directions include west-southwest (WSW) and east-northeast (ENE), although their frequencies are lower compared to SW and NE.
   - The plot suggests that the turbine experiences winds from various directions, but with a clear dominance of southwestern and northeastern winds.

Overall, the plots provide valuable insights into the wind characteristics at the turbine site. The distribution of wind speeds helps identify the most common speed ranges, while the scatter plot confirms the positive relationship between wind speed and energy production. The wind rose plot highlights the prevailing wind directions, which can inform turbine placement and orientation decisions.
